```python
import pandas as pd

df= pd.read_csv("new.csv")

df 
```

```python
df["Price"].mean()
df["Price"].sum()
df["Price"].min()
df["Price"].count()
df["OrderID"].count()
```

```python
df.groupby("Country")["Price"].mean()
df.groupby("Country")["Price"].sum()
df.groupby("Country")["Price"].mean().idxmax()
df.groupby("Country")["Price"].mean().idxmin()
df.groupby("Country")["Price"].mean().sort_values(ascending=False) 
```

```python
df[df["Category"]=="Electronics"]["Price"].mean() 
df[df["Country"]=="USA"]["Price"].sum()
df[df["Category"]=="Electronics"].groupby("Country")["Price"].mean()
df[df["Category"]=="Electronics"].groupby("Country")["Price"].sum().idxmax()
```
